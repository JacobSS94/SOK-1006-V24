```python
import numpy as np
from matplotlib import pyplot as plt
import sympy as sp
```

### Oppgave

*Vi skal her sammenlikne elastisitetene til to lineæare funksjoner. La oss anta at etterspørselen er gitt ved $X_1^D = 500- 5p$ som vi har sett på tidligere, og funnet etterspørselselastisitet $$\epsilon_1^D = -5\frac{p}{x}.$$*      

*Dette skal vi sammenlikne med $X_2^D = 150- p$.*

*(i) Tegn begge etterspørselsfunskjonene i samme figur. Har du noe grunnlag for å si at den ene er "mer/mindre elastisk enn den andre"?*

```python
def etterspørsel1(p):
    return 500-5*p

def etterspørsel2(p):
    return 150-p

p_num=np.linspace(0.1,99.9,100)

def create_ax():
                fig, ax = plt.subplots()
                ax.set_ylabel('Pris (p)', loc='top')
                ax.set_xlabel('Kvantum (x)', loc='right')
                ax.set(xlim=(0,500))
                ax.set(ylim=(0,125))
                ax.spines['top'].set_color('none')
                ax.spines['right'].set_color('none')
                
                return fig, ax

fig, ax = create_ax()

ax.plot(etterspørsel1(p_num),p_num,label='etterspørsel1')
ax.plot(etterspørsel2(p_num),p_num,label='etterspørsel2')

ax.set_title('Lineær etterspørsel, etterspørsel1 og etterspørsel2')
ax.legend(loc='best',frameon=True);
```

* Ved å se på grafene som viser de to etterspørsels funksjonene, vil jeg si den blå linje(etterspørsel1) er  mer elastisk enn den organge(etterspørsel2). For den får større endring i kvantum når prisen reduseres.

*(ii) Tegn to figurer med hver sin etterspørselsfunksjon, og tegn også tilbud $X^S=-20+1.5p$ på begge.*

```python
def etterspørsel1(p):
    return 500-5*p

def etterspørsel2(p):
    return 150-p

def tilbud1 (p):
    return -20+1.5*p

p_num=np.linspace(0.1,99.9,100)

fig, axs = plt.subplots(1,2, sharey=True, figsize=(10,5))

axs[0].plot(etterspørsel1(p_num),p_num,label='etterspørsel1')
axs[0].plot(tilbud1(p_num),p_num,label='tilbud1')
axs[1].plot(etterspørsel2(p_num),p_num,label='etterspørsel2')
axs[1].plot(tilbud1(p_num),p_num,label='tilbud1')

for i in range(2):

    axs[i].set_ylabel('Pris(P)', loc = 'top')
    axs[i].set_xlabel('Kvantum (x)', loc = 'right')
    axs[i].set(xlim=(0,500),ylim=(0,100))
    axs[i].spines['top'].set_color('none')
    axs[i].spines['right'].set_color('none')
    axs[i].legend(loc='best')


fig.suptitle('Tilbud og etterspørsel')

plt.show()
```

*(iii) Anta nå at produksjonskostnader faller slik at tilbudet er nå gitt ved $X^S=-10+1.5p$. Tegn dette inn i begge figurene og undersøk effekten på likevektspris og -kvantum. Hvordan er dette påvirket at etterspørselens priselastisitet? Forklar intuisjonen bak dine funn.*

```python
def etterspørsel1(p):
    return 500-5*p

def etterspørsel2(p):
    return 150-p

def tilbud1(p):
    return -20+1.5*p

def tilbud2(p):
    return -10+1.5*p 

p_num=np.linspace(0.1,99.9,100)

fig, axs = plt.subplots(1,2, sharey=True, figsize=(15,5))


axs[0].plot(etterspørsel1(p_num),p_num,label='etterspørsel1')
axs[0].plot(tilbud1(p_num),p_num,label='tilbud1')
axs[0].plot(tilbud2(p_num),p_num,label='tilbud2')
axs[1].plot(etterspørsel2(p_num),p_num,label='etterspørsel2')
axs[1].plot(tilbud1(p_num),p_num,label='tilbud1')
axs[1].plot(tilbud2(p_num),p_num,label='tilbud2')
for i in range(2):

    axs[i].set_ylabel('Pris(P)', loc = 'top')
    axs[i].set_xlabel('Kvantum (x)', loc = 'right')
    axs[i].set(xlim=(0,500),ylim=(0,100))
    axs[i].spines['top'].set_color('none')
    axs[i].spines['right'].set_color('none')
    axs[i].legend(loc='best')


fig.suptitle('Tilbud og etterspørsel')

plt.show()
```

```python
p, x = sp.symbols('p, x', real=True, positive=True)

eq_s1 = sp.Eq(-20+1.5*p, x)
eq_d1 = sp.Eq(500-5*p, x)

sol_dict = sp.solve ((eq_s1, eq_d1), (p,x))

sol_dict
```
